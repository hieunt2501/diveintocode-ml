```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

[Problem 1] Data acquisition

```python
iris = load_iris()

x = pd.DataFrame(data=iris.data, columns=['sepal_length', 'sepal_width',\
                                          'petal_length', 'petal_width'])
y = pd.DataFrame(data=iris.target, columns=['species'])

print("X =\n{}\nY =\n{}".format(x, y))
```

[Problem 2] Combining data

```python
frames = [x, y]
df = pd.concat(frames, axis=1)
print(df)
```

[Problem 4] Examining the iris dataset itself

The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters

[Problem 6] Creating a diagram

```python
# pie chart of the nmber of samples per label
label = df['species']
firstlabel = label.loc[label == 0]
secondlabel = label.loc[label == 1]
thirdlabel = label.loc[label == 2]
sumlabel = [firstlabel.size, secondlabel.size, thirdlabel.size]

fig = plt.figure()
label = iris.target_names
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
ax.pie(sumlabel, labels=label, autopct='%1.2f%%')
plt.title("Species distribution")
plt.show()
```

```python
# Select features one by one and visualize the distribution of data for each label using a box plot
def drawBoxPlot(data, title,):
    fig, axs = plt.subplots(2, 2)
    axs[0, 0].set_title('Sepal length')
    axs[0, 0].boxplot(data.iloc[:, 0])

    axs[0, 1].set_title('Sepal width')
    axs[0, 1].boxplot(data.iloc[:, 1])

    axs[1, 0].set_title('Petal length')
    axs[1, 0].boxplot(data.iloc[:, 2])

    axs[1, 1].set_title('Petal width')
    axs[1, 1].boxplot(data.iloc[:, 3])

    for ax in axs.flat:
        ax.label_outer()

    fig.suptitle(title)
    plt.show()

setosa = df.loc[df['species'] == 0]
versicolor = df.loc[df['species'] == 1]
virginica = df.loc[df['species'] == 2]

drawBoxPlot(setosa, 'Setosa')
drawBoxPlot(versicolor, 'Versicolor')
drawBoxPlot(virginica, 'Virginica')
```

```python
def drawViolinPlot(data, title):
    fig, axs = plt.subplots(2, 2)
    axs[0, 0].set_title('Sepal length')
    axs[0, 0].violinplot(data.iloc[:, 0])

    axs[0, 1].set_title('Sepal width')
    axs[0, 1].violinplot(data.iloc[:, 1])

    axs[1, 0].set_title('Petal length')
    axs[1, 0].violinplot(data.iloc[:, 2])

    axs[1, 1].set_title('Petal width')
    axs[1, 1].violinplot(data.iloc[:, 3])

    for ax in axs.flat:
        ax.label_outer()

    fig.suptitle(title)
    plt.show()

drawViolinPlot(setosa, 'Setosa')
drawViolinPlot(versicolor, 'Versicolor')
drawViolinPlot(virginica, 'Virginica')
```


*   Box plots: boxplots is great visualization to show a numerical variable. A boxplot shows “four main features about a variable: center, spread, asymmetry, and outliers”
*   Violin plots: are very similar to boxplot. In addition to the four main features, violin plot also shows density of the variable.
Violin plots provide better visualization and information due to its ability to show the density of the variable, hence reveals structure found within the data.


[Problem 7] Confirming the relationship between features

```python
def drawScatter(data, title):
    fig, axs = plt.subplots(1, 2)
    fig.suptitle(title)

    petal_length = data.loc[:, 'petal_length']
    petal_width = data.loc[:, 'petal_width']
    sepal_length = data.loc[:, 'sepal_length']
    sepal_width = data.loc[:, 'sepal_width']

    axs[0].set_title('Petal')
    axs[0].scatter(petal_length, petal_width)
    axs[0].set(xlabel='Length', ylabel='Width')

    axs[1].set_title('Sepal')
    axs[1].scatter(sepal_length, sepal_width)

    plt.xlabel('Length')
    plt.show()

# Take one feature for each of the vertical and horizontal axes and create a color-coded scatter plot for each type (6).
drawScatter(setosa, 'Setosa')
drawScatter(versicolor, 'Versicolor')
drawScatter(virginica, 'Virginica')
```
