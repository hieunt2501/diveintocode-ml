```python
import time
import matplotlib.pyplot as plt

THICKNESS = 0.00008
```

[Problem 1]
Code to calculate the thickness when the paper is folded 43 times

```python
def calThicknessExpo(times):
    folded_thickness = THICKNESS * 2 ** times
    print("Thickness: {} meters.".format(folded_thickness))
    return folded_thickness
```

[Problem 2]
Convert meters to kilometers and display with two decimal places

```python
def convertToKm(thickness):
    print("Thickness: {: .2f} kilometers.".format(thickness/1000))
```

[Problem 3]
Calculating folded thickness using for statement

```python
def calThicknessFor(times):
    folded_thickness = THICKNESS
    for i in range(times):
        folded_thickness *= 2;
    print("Thickness: {}.".format(folded_thickness)) 
    return folded_thickness
```

[Problem 4]
Comparison of calculation time

```python
def calExecTime(f, times):
    start_time = time.time()
    f(times);
    elapsed_tine = time.time() - start_time
    print("Time of function {} : {}[s]".format(f.__name__, elapsed_tine))
```

[Problem 5]
Saving to a list

```python
def calList(times):
    saveList = [THICKNESS]
    for i in range(times):
        saveList.append(saveList[-1] * 2)    
    return saveList    
```

[Problem 6] and [Problem 7]
Displaying a line graph & Customizing graphs

```python
def displayGraph(foldedList, style="-", color="blue"):
    plt.title("Thickness of folded paper")
    plt.xlabel("Number of folds")
    plt.ylabel("Thickness[m]")
    plt.plot(foldedList, style, color=color)
    plt.show()
```

Code executing

```python
print("Thickness using exponential operator")
folded_thickness = calThicknessExpo(43)

print("\nConverting to kilometers")
convertToKm(folded_thickness)

print("\nThickness using for statements")
folded_thickness = calThicknessFor(43)

print("\nExecution time of exponential operator")
calExecTime(calThicknessExpo, 1000)

print("\nExecution time of for statement")
calExecTime(calThicknessFor, 1000)

foldedList = calList(43)
displayGraph(foldedList)
displayGraph(foldedList, 'o', 'g')
displayGraph(foldedList, '--', 'black')
```
