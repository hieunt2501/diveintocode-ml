This is an assignment for solving Sorori Shinzaemon problem from Dive Into Code Machine Learning course

```python
from math import *
from functools import reduce
import numpy as np
import matplotlib.pyplot as plt

# Constant declaration
INITIAL_AMOUNT = 1
GRAINS_PER_DAY = 19400 # According to FreeRice 
PROXIMA_CENTAURI_DISTANCE = 40208 * 10 ** 12

# Functions
def plot_graph(list_grains, list_total_grains, xlabel="grain"):    
    """
    A function that plot graph with list of grains and list of total grains after a certain amount of days.
    Parameters
    --------------
    list_grains: list
        List of grains obtained per day after n days
    list_total_grains: list
        List of total grains obtained per day after n days
    Returns
    --------------
    """
    plt.title("Sorori Shinzaemon Problem")
    plt.xlabel("Number of days")
    plt.ylabel("Rice grain ({})".format(xlabel))
    plt.plot(range(1, len(list_grains) + 1), 
            list_grains, 'o', 
            label='Grains after {} days'.format(len(list_total_grains)))
    plt.plot(range(1, len(list_total_grains) + 1), 
            list_total_grains, 'o', 
            label='Total grains after {} days'.format(len(list_total_grains)))
    plt.legend()
    plt.show()

def compute_sorori_shinzaemon(day=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    # Write the code here. pass is an instruction that means do nothing.
    list_n_grains = [INITIAL_AMOUNT]
    list_total_grains = [list_n_grains[0]]
    for i in range(day - 1):
      list_n_grains.append(list_n_grains[-1] * 2)
      list_total_grains.append(list_total_grains[-1] + list_n_grains[-1])      
    return list_n_grains, list_total_grains

def calDays(total_grains, number_people):
    """
    A function that calculate number of days for a certain number of people could live on with a certain amout of grains
    Parameteres
    --------------
    total_grains: int
        Total rice grains
    number_people: int
        Number of people        
    Returns
    --------------
    Number of days.
    """    
    grains_per_n_person = number_people * GRAINS_PER_DAY
    return floor(total_grains / grains_per_n_person)
```

[Problem 1]: Number of grains on the 100th day

```python
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=100)
print("Total number of grains after {} days is: {}".format(100, list_total_grains[-1]))
plot_graph(list_n_grains, list_total_grains)
```

[Just for fun]
Since the number is too large, we should normalized them by using base 2 logarith for better visualization.

```python
normal_list_n_grains = list(map(lambda x: log(x, 2), list_n_grains[1:]))    
normal_list_total_grains = list(map(lambda x: log(x, 2), list_total_grains[1:]))    
plot_graph(normal_list_n_grains, normal_list_total_grains, "log(2)")
```

[Problem 2] Number of rice grains outside of the 100th day

```python
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=50)
print("Total number of grains after {} days is: {}".format(50, list_total_grains[-1]))
plot_graph(list_n_grains, list_total_grains)
```

[Problem 3]
How many people can live for how many days with the rice we get?

```python
number_people   = 7000000000
total_grains    = list_total_grains[-1]
total_days      = calDays(total_grains,  number_people)
print("With {} grains of rice could feed {} people for {} days.".format(total_grains, number_people, total_days))    
```
