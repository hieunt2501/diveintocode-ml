```python
import numpy as np
import matplotlib.pyplot as plt
```

[Problem 1] Linear function

```python
x = np.arange(-50, 51)
y = 0.5 * x + 1

print('X value:\n{}'.format(x))
print('Y value:\n{}'.format(y))
```

[Problem 2] Array combination

```python
new_array = np.dstack((x, y))[0]
print('New array:\n{}'.format(new_array))
```

[Problem 3] Find the gradient

```python
y_change = y[1:] - y[:-1]
x_change = x[1:] - x[:-1]
grad = y_change/x_change
# dx = x[1] - x[0]
# grad = np.gradient(y, dx)
print(grad.size)
```

[Problem 4] Draw a graph

```python
def drawGraph(xy, grad):
    x, y = xy[:, 0], xy[:, 1]
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function')
    plt.plot(x, y, '-', label="Function")
    plt.plot(x[1:], grad, '-', label="Gradient")
    plt.legend()
    plt.show()
drawGraph(new_array, grad)
```

[Problem 5] Python Functionalization

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    lBound, rBound, interval = x_range
    array_x = np.arange(lBound, rBound, interval, dtype=float)
    array_y = function(array_x)
    array_xy = np.dstack((array_x, array_y))[0]
    gradient = (array_y[1:] - array_y[:-1]) / (array_x[1:] - array_x[:-1])
    return array_xy, gradient

def function1(array_x):
    array_y = array_x ** 2
    return array_y
    
def function2(array_x):
    array_y = 2 * array_x ** 2 + 2 ** array_x
    return array_y

def function3(array_x):
    array_y = np.sin(array_x ** 1/2)
    return array_y

array_xy, gradient = compute_gradient(function1)
drawGraph(array_xy, gradient)
array_xy, gradient = compute_gradient(function2)
drawGraph(array_xy, gradient)
array_xy, gradient = compute_gradient(function3)
drawGraph(array_xy, gradient)
```

[Problem 6] Find the minimum value

```python
def compute_min(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the minimum value using ndarray min and argmin
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    min : int
      Minimum value of function
    """
    lBound, rBound, interval = x_range
    array_x = np.arange(lBound, rBound, interval, dtype=float)
    array_y = function(array_x)
    array_xy = np.dstack((array_x, array_y))[0]
    gradient = (array_y[1:] - array_y[:-1]) / (array_x[1:] - array_x[:-1])
    minIdx = np.argmin(array_y)
    return array_xy, gradient[minIdx-1], gradient[minIdx+1]

array_xy, g1, g2 = compute_min(function1)
print("Slope before and after minimum value: {: .2f} and {: .2f}".format(g1, g2))

array_xy, g1, g2 = compute_min(function2)
print("Slope before and after minimum value: {: .2f} and {: .2f}".format(g1, g2))

array_xy, g1, g2 = compute_min(function3)
print("Slope before and after minimum value: {: .2f} and {: .2f}".format(g1, g2))
```
