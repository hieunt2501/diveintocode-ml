```python
import numpy as np
import matplotlib.pyplot as plt
```

[Problem 1] Creating random numbers

```python
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)
points = np.random.multivariate_normal(mean, cov, 500).reshape(500, 2)
print(points)
```

[Problem 2] Visualization by scatter plot

```python
x = points[:, 0]
y = points[:, 1]
plt.scatter(x, y)
plt.show()
```

[Problem 3] Visualization by histogram

```python
plt.hist(x, density=True)
plt.xlim([-7, 4])
plt.xlabel("x")
plt.ylabel("frequency")
plt.title("X density")
plt.show()

plt.hist(y, density=True)
plt.xlim([-7, 4])
plt.xlabel("y")
plt.ylabel("frequency")
plt.title("Y density")
plt.show()
```

[Problem 4] Addition of data

```python
new_mean = [0, -3]

np.random.seed(0)
new_points = np.random.multivariate_normal(new_mean, cov, 500).reshape(500, 2)
new_x = new_points[:, 0]
new_y = new_points[:, 1]

plt.scatter(x, y, label="0")
plt.scatter(new_x, new_y, label="0")
plt.legend()
plt.show()
```

[Problem 5] Data combination

```python
sum_points = np.vstack((points, new_points))
print(sum_points)
```

[Problem 6] Labeling

```python
tmp = np.zeros((points[:, 1].size, 1))
points = np.concatenate((points, tmp), axis=1)
tmp = np.ones((new_points[:, 1].size, 1))
new_points = np.concatenate((new_points, tmp), axis=1)
sum_points = np.vstack((points, new_points))
print(sum_points)
```
